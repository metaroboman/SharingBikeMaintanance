```python
#python自带csv读取器

def load_content1(add, columns):
    r = []
    add += '/'
    for a in os.listdir(add):
        #print(a)
        if not a.startswith('.'):
            with open(add+a, 'r') as fin:
                [r.append(_) for _ in csv.reader(fin) if _ not in [[], ['', '', '']]]
                #print('%s'%(a))
    #print(r[:20])
    df = pd.DataFrame(r, columns=columns, dtype=np.float)
    #df.dropna(inplace=True, axis=0)
    #df = df[df.n != '']
    
    return df
```

```python
#pandas csv读取器
def load_content2(add, columns):
    r = pd.DataFrame(columns=columns)
    add += '/'
    for a in os.listdir(add):
        if not a.startswith('.'):
            temp = pd.read_csv(add+a, names=columns)
            #print(temp)
            r = pd.concat([r, temp], axis = 0)
    df = r
    df.dropna(inplace=True, axis=0)
    #df = df[df.n != '']
    
    return df
```
