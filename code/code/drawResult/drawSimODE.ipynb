```python
indicators = ['Normal bikes', 'Broken bikes', 'Lost rate', 'Idle rate', 'BP', 'RC', 'DP']
def draw(add, title):
    dfLst = []
    for _add in add:
        _dfLst = [pd.read_csv(a, names=('para', 'mu', 'sig')) for a in _add]
        for df in _dfLst:
            df['low'], df['high'] = df['mu'] - 1.96*df['sig'], df['mu'] + 1.96*df['sig']
        dfLst.append(_dfLst)
    n = int(df.shape[0]/7)
    f, ax = plt.subplots(7,1,figsize=(7,28), sharex=True)
    plt.xlabel('Value')
    #ax[0].set_title(title)
    
    for i in range(7):
        for df in dfLst:
            df0, df1 = df
            re0 = df0[df0.index%7==i]
            re1 = df1[df1.index%7==i]
            ax[i].plot(re1.para, re1.mu,'b')
            ax[i].fill_between(re0.para, re0.high, re0.low, alpha=0.1, color='r')
            ax[i].plot(re0.para, re0.mu, 'r', linestyle='dashed')
            ax[i].set_ylabel(indicators[i])
        #ax[i].legend()
#     f.savefig('/Users/valarian/SJTU/SJTU/毕业论文/article review/My paper/modelAnalysis/graph/central//A50Cen'\
#              +title.replace(' ', '')+'V1.png', dpi = 150, bbox_inches='tight')
```
