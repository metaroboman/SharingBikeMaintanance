```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
indicators = ['Normal bikes', 'Broken bikes', 'Lost rate', 'Idle rate', 'BP', 'RC', 'DP']
def draw(add, title, A, M):
    dfLst = [pd.read_csv(a, names=('para', 'mu', 'sig')) for a in add]
    for df in dfLst:
        df['low'], df['high'] = df['mu'] - 1.96*df['sig'], df['mu'] + 1.96*df['sig']
    n = int(df.shape[0]/7)
    f, ax = plt.subplots(7,1,figsize=(7,28), sharex=True)
    #plt.xlabel('Value')
    #ax[0].set_title(title)
    
    for i in range(7):
        for n, df in enumerate(dfLst):
            re = df[df.index%7==i]
            #ax[i].plot(re1.para, re1.mu, color[0])
            ax[i].fill_between(re.para, re.high, re.low, alpha=0.2)
            marker = ['-', '--', '-.', ':','-', '--', '-.', ':']
            ax[i].plot(re.para, re.mu, linestyle=marker[n])
            if i == 0: ax[0].legend(['Instance'+str(_) for _ in list(range(1,5))])
            ax[i].set_ylabel(indicators[i])
        #ax[i].legend()
    f.savefig('/Users/valarian/SJTU/SJTU/毕业论文/中文论文/graph/sim/'\
             +title.replace(' ', '')+'A'+str(A)+'M'+str(M)+'.png', dpi = 150, bbox_inches='tight')
```
