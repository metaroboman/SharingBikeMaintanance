```python
def addLst(add1, add2):
    lst = [add1+'V'+str(x)+add2 for x in range(4)]
    #lst = [add1 + add2]
    return lst
add1 = '/Users/valarian/SJTU/SJTU/毕业论文/Data/result/PerfIndicators/A2M6'
add2 = '/central/ODE/Beta/varyBeta.csv'
addLst(add1, add2)
```

```python
indicators = ['Normal bikes', 'Broken bikes', 'Lost rate', 'Idle rate', 'BP', 'RC', 'DP']
def draw(add, title):
    dfLst = [pd.read_csv(a, names=('para', 'mu', 'sig')) for a in add]
    for df in dfLst:
        df['low'], df['high'] = df['mu'] - 1.96*df['sig'], df['mu'] + 1.96*df['sig']
    n = int(df.shape[0]/7)
    f, ax = plt.subplots(7,1,figsize=(7,28), sharex=True)
    plt.xlabel('Value')
    #ax[0].set_title(title)
    
    for i in range(7):
        for df in dfLst:
            re = df[df.index%7==i]
            ax[i].plot(re.para, re.mu)
            ax[i].fill_between(re.para, re.high, re.low, alpha=0.1, color='r')
            ax[i].set_ylabel(indicators[i])
        #ax[i].legend()
#     f.savefig('/Users/valarian/SJTU/SJTU/毕业论文/article review/My paper/modelAnalysis/graph/central//A50Cen'\
#              +title.replace(' ', '')+'V1.png', dpi = 150, bbox_inches='tight')
```
